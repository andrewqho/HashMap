```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
```

```python
pd_insert = pd.read_csv('insert.csv')
pd_emplace = pd.read_csv('emplace.csv')
pd_remove = pd.read_csv('remove.csv')
pd_insert_with_removal = pd.read_csv('insert_with_removal.csv')
pd_emplace_with_removal = pd.read_csv('emplace_with_removal.csv')
```

```python
pd_insert.plot(x = 'Num Entries', y = ['Linear Map', 'Robinhood Map', 'Unordered Map'], title="Insert Times, No removals")
pd_emplace.plot(x = 'Num Entries', y = ['Linear Map', 'Robinhood Map', 'Unordered Map'], title="Look-Up Times, No removals")
pd_remove.plot(x = 'Num Entries', y = ['Linear Map', 'Robinhood Map', 'Unordered Map'], title="Removal Times")
pd_insert_with_removal.plot(x = 'Num Entries', y = ['Linear Map', 'Robinhood Map', 'Unordered Map'], title="Insert Times, With Removals")
pd_emplace_with_removal.plot(x = 'Num Entries', y = ['Linear Map', 'Robinhood Map', 'Unordered Map'], title="Look-Up Times, With Removals")
```
